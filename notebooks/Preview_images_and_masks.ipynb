# Предварительное ознакомление с данными

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
```

```python
images_path: str = "../data/JPEGImages"
masks_path: str = "../data/gt_masks"


def load_image(file_path: str) -> np.ndarray:
    return cv2.imread(file_path, cv2.IMREAD_COLOR)


def load_mask(file_path: str) -> np.ndarray:
    return np.load(file_path)


def display_image_and_mask(image: np.ndarray, mask: np.ndarray) -> None:
    fig, axs = plt.subplots(1, 2, figsize=(15, 15))

    axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    axs[0].set_title("Image")

    axs[1].imshow(mask)
    axs[1].set_title("Mask")

    plt.show()
```

# Получим список файлов изображений и масок

```python
if __name__ == "__main__":
    image_files = sorted(os.listdir(images_path))
    mask_files = sorted(os.listdir(masks_path))
```

# Проверим, что количество файлов изображений и масок совпадает

```python
assert len(image_files) == len(mask_files), "Number of images and masks does not match"
```

# Выведем несколько примеров изображений и масок

```python
for i in range(5):
    image_file = os.path.join(images_path, image_files[i])
    mask_file = os.path.join(masks_path, mask_files[i])

    image = load_image(image_file)
    mask = load_mask(mask_file)

    display_image_and_mask(image, mask)
```
